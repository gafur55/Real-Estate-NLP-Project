```python
from transformers import pipeline

# Load a sentiment-analysis pipeline with a pretrained BERT model
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
import pandas as pd

pd.set_option('display.max_colwidth', None)
df = pd.read_csv('processed_zillow_reviews.csv') 

#I'll sample proportionally based on existing data distribution
target_size = 3000

sampled_df = df.groupby('Rating').apply(
    lambda x: x.sample(frac=target_size / len(df), random_state=42)
).reset_index(drop=True)

print(sampled_df['Rating'].value_counts())
sampled_df
```

```python
# Convert cleaned reviews into a list
texts = sampled_df['Cleaned_Description'].tolist()

# Run BERT sentiment prediction (batching optional)
results = sentiment_pipeline(texts)

# Add results to your DataFrame
sampled_df['BERT_Sentiment_Label'] = [res['label'] for res in results]
#This is the confidence score 
sampled_df['BERT_Sentiment_Score'] = [res['score'] for res in results]
```

```python
sampled_df = pd.read_csv('zillow_3k_sample_bert_sentiment.csv')

sampled_df[['Rating', 'Description', 'BERT_Sentiment_Label', 'BERT_Sentiment_Score']]

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Checking the accuracy of Bert with customer's own ratings(1-5)
ct = pd.crosstab(sampled_df['Rating'], sampled_df['BERT_Sentiment_Label'])

sns.heatmap(ct, annot=True, fmt='d', cmap="Blues")
plt.title("BERT Sentiment vs Customer Star Ratings")
plt.xlabel("BERT Predicted Sentiment")
plt.ylabel("Customer Rating (1-5 Stars)")
plt.show()
```
